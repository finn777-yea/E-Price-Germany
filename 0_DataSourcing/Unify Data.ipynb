```python
import pandas as pd

# Carga de archivos CSV
DAX_hourly_2025 = pd.read_csv("DAX_hourly_2025.csv")
Data_weather_2025 = pd.read_csv("Data_weather_2025.csv")
Netztransparenz_2025 = pd.read_csv("Netztransparenz_2025.csv")
SMARTD_2025 = pd.read_csv("SMARTDE_2025.csv")
COMMODITIES_2025 = pd.read_csv("COMMODITIES_2025.csv")
SMARTD_2025 = SMARTD_2025.iloc[:-1]
COMMODITIES_2025 = COMMODITIES_2025.iloc[:-1]
dataframes = {
    "DAX_hourly_2025": DAX_hourly_2025,
    "Data_weather_2025": Data_weather_2025,
    "Netztransparenz_2025": Netztransparenz_2025,
    "smard_generacion_consumo_precios": SMARTD_2025,
    "Commodities": COMMODITIES_2025,
}

for name, df in dataframes.items():
    print(f"DataFrame: {name}")
    print(f"Filas:    {df.shape[0]:,}")
    print(f"Columnas: {df.shape[1]}")
```

```python
# Detectar columna clave (primera columna)
key = DAX_hourly_2025.columns[0]

# Base
df_merged = DAX_hourly_2025.copy()

# Lista de otros dataframes
other_dfs = [
    Data_weather_2025,
    Netztransparenz_2025,
    SMARTD_2025,
    DAX_hourly_2025,
    COMMODITIES_2025
]

for df in other_dfs:
    # eliminar primera columna (asumida como date duplicada)
    df = df.drop(columns=[df.columns[0]])
    
    # concatenar por columnas
    df_merged = pd.concat([df_merged, df], axis=1)

df_merged.columns.values[0] = "date"
print(df_merged.shape)
print(df_merged.head())

# SAVING
df_merged.to_csv("Complete_Data_2025.csv")
```
